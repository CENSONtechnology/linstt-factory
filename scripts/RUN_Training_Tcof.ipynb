# Prepare Input Audio-Text data kaldi for each Database
- Tcof

```bash
## Prepare Input Data for ESTER Phase 1
path_TCOF=$lvcsrRootDir/database/Tcof/Tcof
data_TCOF=$lvcsrRootDir/data/TCOF

## Prepare Audio+Text Input from TCOF
. local/data_prepTCOF.sh $path_TCOF $data_TCOF/data/map_files

## Prepare Input Data from TCOF Text
path_TCOF_text=$lvcsrRootDir/database/Tcof/Tcof_text
data_TCOF_text=$lvcsrRootDir/data/text
./local/text/prepare_text_tcof.sh $path_TCOF_text \
$data_TCOF_text/local/norm/tmp $data_TCOF_text/local/norm/norm_texts \
$data_TCOF_text/local/
```

## Do Phonetic Lexicon Extension

```bash
# Use the new lexicon from UC1 to extend vocabulary with words from ESTER1_2
. RunPhoneticLexiconExtension.sh $data_TCOF_text/local/meeting-vocab.txt $lvcsrRootDir/tools/g2p \
                                 $lvcsrRootDir/training/LEX/LEX0=dict_fr TCOF \
                                 "Generate Words from TCOF based on LEX0=dict_fr" \
                                 12
```

```bash
# For UC1 use the latest vocabulary generated and extend lexicon
. RunPhoneticLexiconExtension.sh $lvcsrRootDir/tmp/words.txt $lvcsrRootDir/tools/g2p \
                                 $lvcsrRootDir/training/LEX/LEX0=dict_frTCOF UC1 "Generate Words from UC1 based on LEX0=dict_frTCOF" 2
                
# Switch Words
. swichTerms.sh $lvcsrRootDir/training/LEX/LEX0\=dict_frTCOFUC1 $lvcsrRootDir/tmp/mapWords.txt \
                $lvcsrRootDir/training/LEX/LEX0=dict_frTCOF
```

## Train Acoustical model

```bash
# Copy the specific data directory that you will use to train acoustical model
# In this example, we use the dir : data/TCOF
echo $data_TCOF
cp -r $data_TCOF $lvcsrRootDir/training/AM/

# Train acoustical model
. trainAcousticModel.sh $lvcsrRootDir/training/AM/TCOF \
                        /opt/lvcsrPlatform/training/LEX/LEX0=dict_frTCOFUC1/lexicon.dict \
                        12
```
